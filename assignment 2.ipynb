# 4.	Create a dictionary with the keys: 'name', 'age', and 'city'. Update the age and add a new key 'email'. Print the final dictionary

```python
# Create a dictionary with keys: 'name', 'age', and 'city'
person = {
    'name': 'Akshata',
    'age': 25,
    'city': 'Jamkhandi'
}

# Print the original dictionary
print("Original dictionary:")
print(person)

# Update the age
person['age'] = 28
print(" updated age:")
print(person)

# Add a new key 'email'
person['email'] = 'akshata@1233.com'

# Print the final dictionary
print("Final dictionary after adding email:")
print(person)
```

# 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- # Use .info(), .describe()
- # Select 'Name' and 'City' columns
- # Drop 'City' column
- # Fill any missing values in 'Salary' column with the mean
- # Remove any duplicate rows


```python
# Create a DataFrame manually
data = {
    'Name': ['Abi', 'Anita', 'Chetana', 'Laxmi', 'Manu', 'Sonu', 'Sanchi', 'Mallu', 'Nandini', 'Sneha'],
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, None],
    'City': ['Pune', 'Ranchi', 'Kolkatta', 'Puri', 'Mumbai', 'Delhi', 'kolkatta', 'ranchi', 'Mumbai', 'Pune'],
    'Salary': [50000, 60000, None, 80000, 90000, 100000, 110000, 120000, None, 140000]
}
df = pd.DataFrame(data)

```

```python
# Use .info() 
df.info()
```

```python
#describe()
df.describe()
```

```python
# Select 'Name' and 'City' columns
name_city = df[['Name', 'City']]
name_city
```

```python
# Drop 'City' column
df_no_city = df.drop('City',axis=1)
df_no_city
```

```python
# Fill missing values in 'Salary' with the mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Salary']
```

```python
# Remove duplicate rows
duplicates = df.drop_duplicates()
duplicates
```
